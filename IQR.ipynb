```python
# Outliner is odd man out
# Lessor range outlier -> is initial stage of outliner ( minimum range)
# Greater range of outlier is last stage of outliner 

# Delete or
# Replace -? replace outliner using IQR  technique (Interquartile Range )
# IQR -> to know the outlier range present in the dataset
# To knpw the range of outliers is IQR -> Q3 - Q1
# Lesser Outlier -> Q1-1.5 * IQR
# Greater outliner -> Q3 + 1.5 * IQR

#Why we are doing 1.5? 

#Reason for the outliers : 1. Human error 2. actual behaviour
```

```python
import pandas as pd
from Univariate import Univariate
```

```python
dataset = pd.read_csv("data.csv")
```

```python
quan, qual = Univariate.QuanQual(dataset)
```

```python
descriptive = Univariate.iqr(dataset, quan)
descriptive
```

```python
lesser =[]
greater =[]
lesser, greater = Univariate.lesserAndGreater(descriptive, quan)
print('lesser' ,lesser)
print ('greater ',greater)
```

```python
dataset = Univariate.applyRanges(dataset, descriptive, lesser, greater )
```

```python
descriptive = Univariate.iqr(dataset, quan)
```

```python
lesser, greater = Univariate.lesserAndGreater(descriptive, quan)
print('lesser' ,lesser)
print ('greater ',greater)
```
